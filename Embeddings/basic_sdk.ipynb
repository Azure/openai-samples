# Python SDK samples

### Get Embeddings
In this example we'll see how to get embeddings using the Azure endpoints.

```python
import json
import openai
import os
```

### Setup

```python
# Load config values
with open(r'../../config.json') as config_file:
    config_details = json.load(config_file)
    
deployment_name = config_details['DEPLOYMENT_NAME']
resource_name = config_details['RESOURCE_NAME']
openai_api_version = config_details['API_VERSION']
```

```python
openai.api_type = "azure"
openai.api_key = os.getenv("OPENAI_API_KEY")
openai.api_base = f"https://{resource_name}.openai.azure.com"
openai.api_version = openai_api_version

embeddings = openai.Embedding.create(engine=deployment_name, input="The food was delicious and the waiter...")["data"][0]["embedding"]
 
# Number of embeddings    
len(embeddings)
```

```python
# Print embeddings
print(embeddings)
```
